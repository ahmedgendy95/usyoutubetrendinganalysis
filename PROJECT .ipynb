```python
import pandas as pd
df = pd.read_csv ("USvideos.csv")
df.head
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.drop(['video_id' , 'thumbnail_link' , 'comments_disabled' ,'ratings_disabled' , 'video_error_or_removed'] , axis=1, inplace=True)
```

```python
df.columns
```

```python
df[['views' , 'likes' , 'dislikes', 'comment_count']].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.histplot(df['views'], bins=50, kde=True)
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['likes'], bins=50, kde=True, color='green')
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='views', y='likes', alpha=0.5)
plt.title('Views vs Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='views', y='comment_count', alpha=0.5, color='orange')
plt.title('Views vs Comment Count')
plt.xlabel('Views')
plt.ylabel('Comment Count')
plt.show()
```

```python
top_channels = df.groupby('channel_title')['views'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_channels.values, y=top_channels.index, palette='viridis')
plt.title('Top 10 Channels with Highest Average Views')
plt.xlabel('Average Views')
plt.ylabel('Channel Title')
plt.show()
```

```python
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
df['trending_day'] = df['trending_date'].dt.day
df['trending_month'] = df['trending_date'].dt.month
```

```python
df['like_ratio'] = df['likes'] / df['views']
```

```python
df['comment_ratio'] = df['comment_count'] / df['views']
```

```python
df[['views', 'likes', 'like_ratio', 'comment_count', 'comment_ratio']].head()
```

```python
threshold = df['views'].quantile(0.80)
threshold
```

```python
df['is_successful'] = df['views'] >= threshold
```

```python
df.groupby('is_successful')['like_ratio'].mean()
```

```python
top_success_channels = df[df['is_successful'] == True]['channel_title'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_success_channels.values, y=top_success_channels.index, palette='magma')
plt.title('Top Channels with Most Successful Videos')
plt.xlabel('Number of Successful Videos')
plt.ylabel('Channel Title')
plt.show()
```

```python
df[df['is_successful'] == True]['trending_month'].value_counts().sort_index().plot(kind='bar', figsize=(10,5))
plt.title('Number of Successful Videos per Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

(الخطوات)

 1. تحميل البيانات وتنظيفها

إستخدمنا dataset عبارة عن الفيديوهات اللي تصدرت الترند وعدد الفيديوهات (الصفوف) أكثر من 40,000  
قمنا بتنظيف البيانات من الأعمدة غير المفيدة مثل الروابط أو الفلاغات الخاصة بإلغاء التعليقات والتقييمات، وكمان اتأكدنا إن البيانات متنسقة بشكل صحيح.


2. معالجة وإعداد البيانات للتحليل

راجعنا القيم المفقودة وتأكدنا من التعامل معاها.  
حوّلنا الأعمدة اللي فيها تواريخ أو أرقام إلى أنواع مناسبة (زي تحويل عمود trending_date إلى نوع datetime).  



 3. تحليل وصفي للأعمدة

عملنا تحليل إحصائي وتصويري لكل عمود على حدة:
- شفنا توزيع المشاهدات، واللايكات، والديسلايكات، وعدد التعليقات.
- استخدمنا الرسوم البيانية لفهم طبيعة انتشار كل متغير وهل فيه outliers أو قيم غريبة.


 4. تحليل العلاقات بين الأعمدة

- حللنا العلاقة بين عدد المشاهدات واللايكات ولقينا علاقة طردية بينهم.
- شفنا العلاقة بين المشاهدات وعدد التعليقات.
- عرفنا أكتر القنوات اللي بتنزل فيديوهات ناجحة وعدد مشاهداتها أعلى من المتوسط.



 5.  إنشاء متغيرات جديدة


علشان نطلع Insights أعمق:
- استخرجنا اليوم والشهر من تاريخ الترند.
- أنشأنا نسبة اللايكات من المشاهدات (Like Ratio).
- أنشأنا نسبة التعليقات من المشاهدات (Comment Ratio).
- جهزنا متغير جديد يحدد هل الفيديو ناجح أو لا بناءً على عدد مشاهداته ضمن أعلى 20%.


 6. تعريف الفيديو الناجح وتحليله

اعتبرنا إن الفيديو الناجح هو اللي عدد مشاهداته من أعلى 20% في البيانات.  
حللنا إيه اللي بيميز الفيديوهات الناجحة:
- عندها Like Ratio وComment Ratio أعلى من غيرها.
- بتظهر أكتر في شهور معينة.
- بعض القنوات عندها معدل إنتاج فيديوهات ناجحة عالي جدًا.


 أهم النتائج:

- ممكن التنبؤ باحتمالية نجاح فيديو من خلال بياناته فقط، زي اللايكات، وقت النشر، ونشاط القناة.
